```
import pandas as pd
import numpy as np

import ast

# Corrected Code Fixing the KeyError and malformed node or string for 'behavior_post'

# Check for non-string or invalid behavior_post values and handle them
def safe_literal_eval(val):
    try:
        if isinstance(val, str):
            return ast.literal_eval(val)
        elif isinstance(val, list):
            return val
        else:
            return []  # Return an empty list if the format is invalid
    except (ValueError, SyntaxError):
        return []  # Return an empty list in case of an error

# Aggregate Perception and Belief Scores
final_df['belief_pre_score'] = final_df[['belief_pre_1', 'belief_pre_2', 'belief_pre_3', 'belief_pre_4']].mean(axis=1)
final_df['belief_post_score'] = final_df[['belief_post_1', 'belief_post_2', 'belief_post_3', 'belief_post_4']].mean(axis=1)
final_df['belief_change'] = final_df['belief_post_score'] - final_df['belief_pre_score']

# Aggregate Distance Perception Scores
final_df['distance_pre_score'] = final_df[['distance_pre_1', 'distance_pre_2', 'distance_pre_3', 'distance_pre_4', 'distance_pre_5']].mean(axis=1)
final_df['distance_post_score'] = final_df[['distance_post_1', 'distance_post_2', 'distance_post_3', 'distance_post_4', 'distance_post_5']].mean(axis=1)
final_df['distance_change'] = final_df['distance_post_score'] - final_df['distance_pre_score']

# Aggregate Policy Support
final_df['policy_pre_score'] = final_df[['policy_pre_1', 'policy_pre_2', 'policy_pre_3', 'policy_pre_4', 'policy_pre_5', 'policy_pre_6', 'policy_pre_7']].mean(axis=1)
final_df['policy_post_score'] = final_df[['policy_post_1', 'policy_post_2', 'policy_post_3', 'policy_post_4', 'policy_post_5', 'policy_post_6', 'policy_post_7']].mean(axis=1)
final_df['policy_change'] = final_df['policy_post_score'] - final_df['policy_pre_score']

# Aggregate Product Adoption
final_df['adoption_pre_score'] = final_df[['adoption_pre_1', 'adoption_pre_2', 'adoption_pre_3', 'adoption_pre_4', 'adoption_pre_5', 'adoption_pre_6']].mean(axis=1)
final_df['adoption_post_score'] = final_df[['adoption_post_1', 'adoption_post_2', 'adoption_post_3', 'adoption_post_4', 'adoption_post_5', 'adoption_post_6']].mean(axis=1)
final_df['adoption_change'] = final_df['adoption_post_score'] - final_df['adoption_pre_score']

# Aggregate General Sharing 
# Recode `general_sharing_pre` based on `sharing_pre`
final_df['general_sharing_pre'] = final_df['sharing_pre'].apply(
    lambda x: 1 if x == 1 else (0 if x == 2 else None)
)
# Recode `general_sharing_post` based on `sharing_post`
final_df['general_sharing_post'] = final_df['sharing_post'].apply(
    lambda x: 1 if x == 1 else (0 if x == 2 else None)
)
final_df['general_sharing_change'] = final_df['general_sharing_post'] - final_df['general_sharing_pre']

# Aggregate Intervention Sharing
final_df['intervention_sharing'] = final_df['sharingwilling_post'].apply(
    lambda x: 1 if x == 1 else (0 if x == 2 else None)
)

# Aggregate single_use2 and refillable together as Intention
final_df['intention_pre_score'] = final_df[['single_use2_pre', 'refillable_pre']].mean(axis=1)
final_df['intention_post_score'] = final_df[['single_use2_post', 'refillable_post']].mean(axis=1)
final_df['intention_change'] = final_df['intention_post_score'] - final_df['intention_pre_score']

# Aggregate the Mean of single_use_pre and single_use_post
final_df['single_use_perception'] = final_df[['single_use_pre', 'single_use_post']].mean(axis=1)

animals = ["seahorse", "beluga", "jellyfish"]


# Define a function to extract the animal name from the group
def extract_animal(group_name):
    for animal in animals:  # Use the predefined list of animals
        if animal in group_name:
            return animal
    return "Unknown"  # Return "Unknown" if no animal is found in the group name

# Apply the function to create the animal column
final_df['animal'] = final_df['group'].apply(extract_animal)

# Display the first few rows to verify the new column
print(final_df[['group', 'animal']].head())


# Define the treatment and control groups
control_group = [
    "control_scientific_seahorse", 
    "control_scientific_beluga",
    "control_scientific_jellyfish",
    "control_personal_seahorse", 
    "control_personal_beluga",
    "control_personal_jellyfish"
    ]
treatment_group = [
    "treatment_seahorse",
    "treatment_beluga",
    "treatment_jellyfish"
    ]

# Define the list of possible animals
animals = ["seahorse", "beluga", "jellyfish"]

# Define a function to extract the animal name from the group
def extract_animal(group_name):
    for animal in animals:
        if isinstance(group_name, str) and animal in group_name:  # Check if animal exists in the group name
            return animal
    return "Unknown"  # Default to "Unknown" if no animal is found

# Apply the function to create the animal column
final_df['animal'] = final_df['group'].apply(extract_animal)

# Verify the output
print(final_df[['group', 'animal']].head())

# Remove any row with group label not in the treatment or control group
final_df = final_df[final_df['group'].isin(control_group + treatment_group)]

# Define mapping for categories
group_category_mapping = {
    "control_scientific_seahorse": "control_scientific",
    "control_scientific_beluga": "control_scientific",
    "control_scientific_jellyfish": "control_scientific",
    "control_personal_seahorse": "control_personal",
    "control_personal_beluga": "control_personal",
    "control_personal_jellyfish": "control_personal",
    "treatment_seahorse": "treatment",
    "treatment_beluga": "treatment",
    "treatment_jellyfish": "treatment",
}

# Add the `group_category` column
final_df['group_category'] = final_df['group'].map(group_category_mapping)

# Ensure no missing categories are left undefined
final_df['group_category'] = final_df['group_category'].fillna('Unknown')

# Define the group-to-label mapping
group_labels = {
    "control_scientific_seahorse": "Scientific Fact + Image",
    "control_scientific_beluga": "Scientific Fact + Image",
    "control_scientific_jellyfish": "Scientific Fact + Image",
    "control_personal_seahorse": "Personal Story + Character",
    "control_personal_beluga": "Personal Story + Character",
    "control_personal_jellyfish": "Personal Story + Character",
    "treatment_seahorse": "Chatbot + Character",
    "treatment_beluga": "Chatbot + Character",
    "treatment_jellyfish": "Chatbot + Character",
}

# Add the 'group_label' column
final_df['group_label'] = final_df['group'].map(group_labels)

# Introduce a new 'condition' column
final_df['condition'] = final_df['group'].apply(
    lambda x: 'treatment' if x in treatment_group else ('control' if x in control_group else 'Unknown')
)

# Convert from string -> actual Python list
final_df['behavior_post'] = final_df['behavior_post'].apply(safe_literal_eval)

# Convert each string inside that list -> int
final_df['behavior_post'] = final_df['behavior_post'].apply(lambda x: [int(item) for item in x])

# Define sustainable and non-sustainable products
sustainable_products = [1, 5, 7, 9, 11, 13]
non_sustainable_products = [4, 6, 8, 10, 12, 14]

# Create sustsainable_choices and non_sustainable_choices columns
# TODO: Add consumption Ratio
final_df['sustainable_choices'] = final_df['behavior_post'].apply(
    lambda x: [p for p in x if p in sustainable_products]
)
final_df['non_sustainable_choices'] = final_df['behavior_post'].apply(
    lambda x: [p for p in x if p in non_sustainable_products]
)

# Create ratio of sustainable choices against non-sustainable choices
# Define the function to compute your custom ratio
def custom_ratio(sustainable_list, non_sustainable_list):
    count_sustainable = len(sustainable_list)
    count_non_sustainable = len(non_sustainable_list)
    
    # Case 1: Both counts exist - compute the ratio
    if count_sustainable > 0 and count_non_sustainable > 0:
        return count_sustainable / count_non_sustainable
    # Case 2: Only sustainable products are chosen (non_sustainable is zero)
    elif count_non_sustainable == 0 and count_sustainable > 0:
        return count_sustainable
    # Case 3: Only non_sustainable products are chosen (sustainable is zero)
    elif count_sustainable == 0 and count_non_sustainable > 0:
        return 0
    # Case 4: Both are zero; you may decide to return NaN or another value
    else:
        return np.nan

# Apply the function row-wise to create the new column
final_df['sustainable_choices_ratio'] = final_df.apply(
    lambda row: custom_ratio(row['sustainable_choices'], row['non_sustainable_choices']),
    axis=1
)
```

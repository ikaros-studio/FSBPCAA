```
import pandas as pd
import scipy.stats as stats
import numpy as np
import plotly.express as px
from statannotations.Annotator import Annotator
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.collections import PathCollection

import seaborn as sns
import sys
import scipy
sys.path.append('/work/include/')
import stat_process
import importlib

import random

# Define a function to balance the groups, if needed
def balance_groups(final_df, group_column='group'):
    """
    Ensures an even number of participants in each group by randomly reducing
    the size of overrepresented groups.
    
    Parameters:
        final_df (DataFrame): The dataframe to process.
        group_column (str): The column indicating participant groups.
    
    Returns:
        DataFrame: A dataframe with balanced group sizes.
    """
    group_counts = final_df[group_column].value_counts()
    min_group_size = group_counts.min()  # Determine the smallest group size

    # Balanced dataframe to hold participants
    balanced_final_df = pd.DataFrame()
    
    # Adjust each group to the minimum size
    for group, count in group_counts.items():
        group_data = final_df[final_df[group_column] == group]
        if count > min_group_size:
            # Randomly sample participants to match the smallest group size
            group_data = group_data.sample(min_group_size, random_state=42)
        balanced_final_df = pd.concat([balanced_final_df, group_data])
    
    return balanced_final_df


importlib.reload(stat_process)

# Define experiment constants and validation rules
EXPERIMENT_CONFIG = {
    'target_n_per_condition': 400,
    'total_target_n': 1200,
    'min_completion_time_seconds': 180,  # 3 minutes minimum
}

# Create validation functions
def validate_completion_time(final_df):
    """Check if completion time meets minimum threshold"""
    duration = pd.to_numeric(final_df['Duration (in seconds)'], errors='coerce')
    return duration >= EXPERIMENT_CONFIG['min_completion_time_seconds']

def validate_attention_checks(final_df):
    """Verify attention checks are passed"""
    pre_check = pd.to_numeric(final_df['attention_pre'], errors='coerce') == 1
    post_check = pd.to_numeric(final_df['attention_post'], errors='coerce') == 4
    return pre_check & post_check

def validate_response_patterns(final_df):
    """Check for straight-lining and other suspicious patterns"""
    # Add specific pattern checks based on your survey structure.
    return pd.Series([True] * len(final_df))  # Currently, simply pass all participants

# --- Outlier Detection Functions ---

# def detect_outliers(final_df, variables, z_threshold=3):
#     """
#     Returns a boolean Series with True for rows that are NOT outliers.
#     A row is flagged as an outlier in a given column if its z-score exceeds z_threshold.
#     The function requires the variables list to consist of numeric columns.
#     """
#     valid = pd.Series(True, index=final_df.index)
#     for var in variables:
#         if var in final_df.columns and pd.api.types.is_numeric_dtype(final_df[var]):
#             z = np.abs((final_df[var] - final_df[var].mean()) / final_df[var].std())
#             valid &= (z <= z_threshold)
#     return valid

# # List of key variables for outlier detection on 0 to 100 scale.
# outlier_vars = [
#     'belief_pre_score', 'belief_post_score',
#     'distance_pre_score', 'distance_post_score',
#     'policy_pre_score', 'policy_post_score',
#     'adoption_pre_score', 'adoption_post_score'
# ]

# Create a mask for technical issues
def validate_technical_issues(final_df):
    """Exclude participants with reported technical issues."""
    return final_df['issues'] != 1

# Create exclusion tracking (now including outlier count)
def track_exclusions(final_df):
    """Track reasons for excluding participants, including technical issues."""
    failed_time = (~validate_completion_time(final_df)).sum()
    failed_attention = (~validate_attention_checks(final_df)).sum()
    failed_patterns = (~validate_response_patterns(final_df)).sum()
    # failed_outliers = (~detect_outliers(final_df, outlier_vars, z_threshold=3)).sum()
    failed_technical_issues = (final_df['issues'] == 1).sum()  # Count participants with technical issues
    
    exclusions = {
        'total_responses': len(final_df),
        'failed_completion_time': failed_time,
        'failed_attention_check': failed_attention,
        'suspicious_patterns': failed_patterns,
        # 'failed_outlier_check': failed_outliers,
        'technical_issues': failed_technical_issues,  # Log technical issues
        'total_excluded': failed_time + failed_attention + failed_patterns + failed_technical_issues # Readd failed_outliers
    }
    exclusions['final_sample'] = exclusions['total_responses'] - exclusions['total_excluded']
    return pd.Series(exclusions)

# Monitor condition assignment
def monitor_condition_balance(final_df):
    """Track participant counts per condition"""
    condition_counts = final_df['group'].value_counts()
    target_n = EXPERIMENT_CONFIG['target_n_per_condition']
    
    balance_status = pd.DataFrame({
        'current_n': condition_counts,
        'target_n': pd.Series([target_n] * len(condition_counts), index=condition_counts.index),
        'remaining_n': pd.Series([target_n] * len(condition_counts), index=condition_counts.index) - condition_counts
    })
    return balance_status

# Print exclusion summary BEFORE removal of participants:
print("\nExclusion Summary (Before Filtering):" + "\n")
print(track_exclusions(final_df))

# Remove exclusions from dataframe:
# Here we combine all validation criteria, including outlier removal.
valid_mask = (validate_completion_time(final_df) &
              validate_attention_checks(final_df) &
              validate_response_patterns(final_df) &
              # detect_outliers(final_df, outlier_vars, z_threshold=3) &
              validate_technical_issues(final_df) 
              )  
final_df = final_df[valid_mask]

# If you need to balance the groups, do it here:
# final_df = balance_groups(final_df)
print("\nCondition Balance (After Filtering):")
print(monitor_condition_balance(final_df))
```

```
import pandas as pd
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import scipy
from matplotlib.collections import PathCollection

# If you're using a Jupyter notebook, uncomment the following line:
# %matplotlib inline

# Append custom paths if necessary
sys.path.append('/work/include/')
import stat_process
import importlib
importlib.reload(stat_process)

# --------------------------------------------
# Experiment Configuration
# --------------------------------------------
EXPERIMENT_CONFIG = {
    'conditions': {
        "chat_nonpersonal_action": "Chatbot (Non-Personal) + Action",
        "chat_nonpersonal_control": "Chatbot (Non-Personal) + Control",
        "chat_nonpersonal_future": "Chatbot (Non-Personal) + Future",
        "chat_nonpersonal_moral": "Chatbot (Non-Personal) + Moral",
        "chat_personalized_action": "Chatbot (Personalized) + Action",
        "chat_personalized_control": "Chatbot (Personalized) + Control",
        "chat_personalized_future": "Chatbot (Personalized) + Future",
        "chat_personalized_moral": "Chatbot (Personalized) + Moral",
        "text_statement_action": "Text Statement + Action",
        "text_statement_control": "Text Statement + Control",
        "text_statement_future": "Text Statement + Future",
        "text_statement_moral": "Text Statement + Moral"
    },
    'dependent_vars': {
        "belief_change": 'Belief Change',
        "policy_adoption_change": 'Policy Change',
        "general_sharing_change": "Sharing Change",
        "consumption_change": 'Consumption Change',
        "distance_change": 'Distance Change',
        "intentions_change": 'Intentional Change',
        "choices": 'Sustainable Choices',
    }
}

# --------------------------------------------
# Function to Compute Cohen's d
# --------------------------------------------
def cohen_d(x, y):
    """
    Compute Cohen's d for two independent samples.
    x: baseline group values (array-like)
    y: comparison group values (array-like)
    """
    nx = len(x)
    ny = len(y)
    # Compute sample standard deviations (use ddof=1 for sample std)
    sx = np.std(x, ddof=1)
    sy = np.std(y, ddof=1)
    # Compute pooled standard deviation
    pooled_sd = np.sqrt(((nx - 1) * sx**2 + (ny - 1) * sy**2) / (nx + ny - 2))
    if pooled_sd == 0:
        return 0.0
    return (np.mean(y) - np.mean(x)) / pooled_sd

# --------------------------------------------
# Load and Prepare Data
# --------------------------------------------
# Define the path to your data
data_path = "/work/processed/cleaned_data.csv"

# Load your data
try:
    data = pd.read_csv(data_path, header=0)
except FileNotFoundError:
    raise FileNotFoundError(f"The file at {data_path} was not found.")
except Exception as e:
    raise Exception(f"An error occurred while reading the file: {e}")

# Ensure that 'combined_condition' column exists
if 'combined_condition' not in data.columns:
    raise KeyError("The dataset does not contain a 'combined_condition' column.")

# --------------------------------------------
# Compute Effect Sizes and P-Values
# --------------------------------------------
baseline = "text_statement_control"
summary_list = []

# Verify that the baseline condition exists in the data
if baseline not in data['combined_condition'].unique():
    raise ValueError(f"The baseline condition '{baseline}' is not present in the data.")

# Loop over each dependent variable and compute effect sizes and p-values.
for dvar, friendly_label in EXPERIMENT_CONFIG['dependent_vars'].items():
    # Extract baseline group values for this variable.
    baseline_values = data.loc[data['combined_condition'] == baseline, dvar].dropna()
    
    if baseline_values.empty:
        print(f"Warning: No data found for baseline condition '{baseline}' in dependent variable '{dvar}'. Skipping.")
        continue
    
    # Prepare a list to store effect sizes and p-values for each condition.
    effect_entries = []
    
    # Iterate through each unique group in the dataset.
    for grp in data['combined_condition'].unique():
        group_values = data.loc[data['combined_condition'] == grp, dvar].dropna()
        
        if grp == baseline:
            effect = 0.0
            p_value = np.nan  # No comparison with itself
        else:
            if group_values.empty:
                print(f"Warning: No data found for condition '{grp}' in dependent variable '{dvar}'. Setting effect_size and p_value to NaN.")
                effect = np.nan
                p_value = np.nan
            else:
                # Compute Cohen's d
                effect = cohen_d(baseline_values, group_values)
                # Compute p-value using Welch's t-test
                try:
                    ttest = stats.ttest_ind(baseline_values, group_values, equal_var=False)
                    p_value = ttest.pvalue
                except Exception as e:
                    print(f"Warning: T-test failed for condition '{grp}' in dependent variable '{dvar}': {e}")
                    p_value = np.nan
        
        # Get the friendly condition name.
        cond_friendly = EXPERIMENT_CONFIG["conditions"].get(grp, grp)
        
        effect_entries.append({
            "combined_condition": grp,
            "condition": cond_friendly,
            "dependent_var": friendly_label,
            "effect_size": effect,
            "p_value": p_value
        })
    
    # Append this dependent variable's results to the summary list.
    summary_list.append(pd.DataFrame(effect_entries))

# Combine all effect size summaries into one DataFrame.
summary_df = pd.concat(summary_list, ignore_index=True)

# OPTION B: Remove the baseline from the final plot.
summary_df = summary_df[summary_df['combined_condition'] != baseline]

# --------------------------------------------
# Define a Comprehensive Color Map
# --------------------------------------------

# Define colors for each condition, grouped by their main category
color_map = {
    # Chatbot (Non-Personal) + X
    "Chatbot (Non-Personal) + Action": "#1f77b4",    # Blue
    "Chatbot (Non-Personal) + Control": "#aec7e8",   # Light Blue
    "Chatbot (Non-Personal) + Future": "#657b83",    # Slate Blue
    "Chatbot (Non-Personal) + Moral": "#08306b",     # Dark Blue
    
    # Chatbot (Personalized) + X
    "Chatbot (Personalized) + Action": "#2ca02c",    # Green
    "Chatbot (Personalized) + Control": "#98df8a",   # Light Green
    "Chatbot (Personalized) + Future": "#225ea8",    # Teal
    "Chatbot (Personalized) + Moral": "#006d2c",     # Dark Green
    
    # Text Statement + X
    "Text Statement + Action": "#d62728",            # Red
    "Text Statement + Control": "#ff9896",           # Light Red
    "Text Statement + Future": "#e377c2",            # Pink
    "Text Statement + Moral": "#c7c7c7",             # Grey
}

# Verify that all conditions are included in the color_map
unique_conditions = summary_df['condition'].unique()
missing_conditions = set(unique_conditions) - set(color_map.keys())
if missing_conditions:
    print(f"Warning: The following conditions are missing in the color_map: {missing_conditions}")
    # Optionally, assign default colors to missing conditions
    default_palette = sns.color_palette("husl", len(missing_conditions))
    for cond, color in zip(missing_conditions, default_palette):
        color_map[cond] = sns.utils.rgb2hex(color)

# --------------------------------------------
# Verify Color Mapping
# --------------------------------------------
# Check if all conditions have a color assigned
assert all(cond in color_map for cond in unique_conditions), "Not all conditions have a color assigned in color_map."

# --------------------------------------------
# Plotting: Create a Grouped Bar Chart with Expanded Axes
# --------------------------------------------
plt.figure(figsize=(20, 8))  # Increased figure width and height

# Create the barplot using 'condition' as hue and applying the color_map
ax = sns.barplot(
    data=summary_df,
    x='dependent_var',
    y='effect_size',
    hue='condition',
    palette=color_map,  # Apply the defined color map
    order=dep_order,
    dodge=True  # Creates grouped bars
)

# Annotate each bar with its effect size value on top.
for p in ax.patches:
    height = p.get_height()
    if pd.isna(height):
        continue  # Skip if height is NaN
    x_center = p.get_x() + p.get_width() / 2
    ax.text(
        x_center, height + 0.02, f'{height:.2f}',
        ha='center', va='bottom',
        fontsize=10, fontweight='bold'
    )

# Increase the y-axis range with a bit more margin.
ymin, ymax = ax.get_ylim()
ax.set_ylim(ymin - 0.1, ymax + 0.1)

# Now annotate each bar with its corresponding p-value at the bottom.
# Only annotate if p_value is not NaN and p_value < 0.1
offset = 0.02 * (ymax - ymin)  # Adjust annotation offset based on the y-axis range
for i, p in enumerate(ax.patches):
    # Retrieve the corresponding p-value from the summary DataFrame.
    row = summary_df.reset_index(drop=True).iloc[i]
    p_val = row['p_value']
    
    # Ensure p_val is a float
    try:
        p_val = float(p_val)
    except (ValueError, TypeError):
        p_val = np.nan
    
    # Only annotate if p_value is not NaN and p_value < 0.1
    if not pd.isna(p_val) and p_val < 0.1:
        # Format the p-value: values below 0.001 are shown as 'p<0.001'
        if p_val < 0.001:
            p_text = "p<0.001"
        else:
            p_text = f"p={p_val:.3f}"
        
        # Determine the vertical coordinate for the annotation.
        if p.get_height() >= 0:
            y_pos = 0 - offset
        else:
            y_pos = (p.get_y() + p.get_height()) - offset

        x_center = p.get_x() + p.get_width() / 2
        ax.text(
            x_center, y_pos, p_text,
            ha='center', va='top', fontsize=10, color='black', fontweight='bold'
        )
    else:
        # Debugging statement to confirm p-value is not being annotated
        if not pd.isna(p_val):
            print(f"Skipping annotation for condition '{row['condition']}' with p-value: {p_val}")
        else:
            print(f"Skipping annotation for condition '{row['condition']}' due to NaN p-value.")

# Labels, title, and layout improvements.
plt.xlabel('Dependent Variable', fontsize=14, fontweight='bold')
plt.ylabel("Effect Size (Cohen's d)", fontsize=14, fontweight='bold')
plt.title(
    "Combined Summary of Effect Sizes Across Conditions\n(Relative to Control)", 
    fontsize=16, fontweight='bold'
)
plt.xticks(rotation=45)
plt.legend(title="Condition", title_fontsize='13', fontsize='11', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Display the plot
plt.show()
```
